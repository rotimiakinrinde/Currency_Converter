```python
import os
from dotenv import load_dotenv
import requests
import json
import pandas as pd
```

```python
# load environment variables from .env file
load_dotenv()

api_key = os.getenv("EXCHANGE_API_KEY")

print(f"API Key Loaded: {'Yes!!!' if api_key else 'NO!!!'}")
```

```python
base_url = f"https://api.exchangeratesapi.io/v1/latest"

# Query parameters (key included here only for request)
params = {
    'access_key': api_key
    }
```

```python
response = requests.get(base_url, params=params)

# Parse and pretty-print JSON
if response.status_code == 200:
    data = response.json()
    print(json.dumps(data, indent=4))
else:
    print(f"Error {response.status_code}:")
    print(json.dumps(response.json(), indent=4))
```

```python
# gather the parameters of interest
# construct the URl and send a Get request to it
# For unsuccessful request: print the error message
# For successful request: extract the data and calculate the result
# Display the result to the user



# Base URL

exchange_url = "https://api.exchangeratesapi.io/v1"

base_url = exchange_url


# Gather input from user
date = input("Please enter the date (in the format 'yyyy-mm-dd' or 'latest'): ")
base_currency = input("Convert from (currency): ").upper()
target_currency = input("Convert to (currency): ").upper()
quantity = float(input(f"How much {base_currency} do you want to convert: "))

# Prepare request parameters
params = {
    'access_key': api_key,
    'base': base_currency,
    'symbols': target_currency
}

# Make the request
url = f"{base_url}/{date}"
response = requests.get(url, params=params)

# Handle response
if not response.ok:
    print(f"\nError {response.status_code}:")
    print(response.json().get('error', 'Unknown error'))
else:
    data = response.json()
    rate = data['rates'][target_currency]
    result = quantity * rate
    print(f"\n{quantity} {base_currency} = {result:.4f} {target_currency} on {data['date']}")
```
